# Demo Gradient Descent (Manual)

```python
def poly(x):
  return (2*x*x*x*x) + (x*x*x) - (3*x*x)

x = np.linspace(-2, 2, 40)
print(x)
plt.plot(x, poly(x))
plt.show()
```

```python
def deriv_poly(x):
  return (8*x*x*x) + (3*x*x) - (6*x)

EPOCH = 500
ALPHA = 0.001
xs    = [0.] * (EPOCH + 1)
xs[0] = 2.0
for t in range(EPOCH):
  xs[t + 1] = xs[t] - ALPHA * deriv_poly(xs[t])

print("optimum at x = {} ---> y = {}".format(xs[-1], poly(xs[-1])))

x = np.linspace(-2, 2, 40)
xs = np.array(xs)
plt.plot(x, poly(x), color = 'blue')
plt.plot(xs, poly(xs), color = 'red', linestyle = "None", marker = 'o')
plt.show()
```



# Gradient Descent dengan Tensorflow

```python
import tensorflow as tf
```

```python
def poly(x):
  return (2*x*x*x*x) + (x*x*x) - (3*x*x)
```

## Manual

```python
EPOCH = 100
ALPHA = 0.01

# definisikan sebuah variable
x = tf.Variable(2.0)

for i in range(EPOCH):

  with tf.GradientTape() as tape:
    y = poly(x)

  grads = tape.gradient(y, [x])
  print ('y = {:.3f}, x = {:.3f},  grads = {:.3f}'.format(y.numpy(), x.numpy(), grads[0].numpy()))
  x.assign(x - ALPHA * grads[0].numpy())
```

## Dengan Modul Optimizer

```python
EPOCH = 100
ALPHA = 0.01

# SGD optimizer
opt = tf.keras.optimizers.SGD(learning_rate = ALPHA)
#opt = tf.keras.optimizers.Adam(learning_rate = ALPHA)

# definisikan sebuah variable
x = tf.Variable(2.0)

for i in range(EPOCH):

  with tf.GradientTape() as tape:
    y = poly(x)

  grads = tape.gradient(y, [x])
  processed_grads = [g for g in grads]
  grads_and_vars = zip(processed_grads, [x])

  print ('y = {:.3f}, x = {:.3f},  grads = {:.3f}'.format(y.numpy(), x.numpy(), grads[0].numpy()))

  opt.apply_gradients(grads_and_vars)
```

# Gradient Descent dengan PyTorch

```python
import torch

from torch.autograd import Variable # deprecated
```

```python
def poly(x):
  return (2*x*x*x*x) + (x*x*x) - (3*x*x)

EPOCH = 100
ALPHA = 0.01

#x = Variable(torch.tensor(2.0), requires_grad = True)
x = torch.tensor(2.0, requires_grad = True)

for i in range(EPOCH):

  y = poly(x)

  y.backward() # hitung gradient
  # Since PyTorch accumulates gradients by default, if you were to call y.backward() 
  # multiple times without resetting x.grad, the gradients from each backward call 
  # would accumulate. 
  with torch.no_grad(): # operasi di dalam with tidak akan track gradient
    # torch.no_grad() context is used to prevent certain operations 
    # (like the gradient descent update step) from being added to the computation graph. 
    # If the update step were tracked in the graph, PyTorch would attempt to differentiate it in future iterations, 
    # leading to incorrect gradients and an inefficiently large graph.
    grad = x.grad
    x -= grad * ALPHA
    print ('y = {:.3f}, x = {:.3f},  grad = {:.3f}'.format(y.data, x.data, grad.data))
    x.grad.zero_() # set gradient kembali ke nol
    # PyTorch accumulates gradients by default, meaning that each time you call .backward(), 
    # the computed gradient is added to any existing values in the .grad attribute rather than overwriting it. 
    # This behavior can be useful in specific scenarios, 
    # especially when working with mini-batch processing 
    # or accumulating gradients across multiple backward passes.
```

# Contoh dengan 2 Trainable Variables

Coba cari nilai $x,y$ sehingga fungsi $f(x,y) = x^2 + y^2 + 2x - 4y + 5$ mencapai local minima!

```python
def f(x, y):
  return (x*x) + (y*y) + (2*x) - (4*y) + 5

def dfdx(x):
  return 2*x + 2

def dfdy(y):
  return 2*y - 4

EPOCH = 500
ALPHA = 0.01
xs = [0.] * (EPOCH + 1)
ys = [0.] * (EPOCH + 1)
xs[0], ys[0] = -2.0, 2.0
for t in range(EPOCH):
  xs[t + 1] = xs[t] - ALPHA * dfdx(xs[t])
  ys[t + 1] = ys[t] - ALPHA * dfdy(ys[t])

print("optimum at x = {}, y = {} dengan nilai f(x,y) = {}".format(xs[-1], ys[-1], f(xs[-1], ys[-1])))
```

```python
# inaya
plt.figure()
xs = np.linspace(-5, 5, 100)
ys = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(xs, ys)

cp = plt.contour(X, Y, f(X, Y), cmap='viridis')
plt.colorbar(cp)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour plot of f(x, y)')
plt.show()
```

# Latihan

Coba cari nilai $x,y$ sehingga fungsi $f(x,y) = 50 - x^2 - 2y^2$ mencapai local maximum!

```python
def f(x,y):
    return 50 - x*x - 2*y*y


def dfdx(x):
  return -2*x

def dfdy(y):
  return -4*y

EPOCH = 500
ALPHA = 0.01
xs = [0.] * (EPOCH + 1)
ys = [0.] * (EPOCH + 1)
xs[0], ys[0] = -2.0, 2.0
for t in range(EPOCH):
  xs[t + 1] = xs[t] + ALPHA * dfdx(xs[t])
  ys[t + 1] = ys[t] + ALPHA * dfdy(ys[t])

print("optimum at x = {}, y = {} dengan nilai f(x,y) = {}".format(xs[-1], ys[-1], f(xs[-1], ys[-1])))

```
